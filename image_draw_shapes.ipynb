# Display Images and draw basic shapes over them

**Oper an image and display it in correct RGB order**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("DATA/dog_backpack.png")
```

```python
plt.imshow(img)
```

**Fix RGB order**

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

**Rotate an image upside down**

```python
new_img = img_rgb.copy()
```

```python
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

## Draw a Red rectangle aroud the dog face

**Points of the rectangle**

```python
pt1 = (200,380)
pt2 = (600,700)
```

**Draw the rectangle in red color with thickness param**

```python
cv2.rectangle(img_rgb,pt1=pt1,pt2=pt2,color=(255,0,0),thickness=10)
plt.imshow(img_rgb)
```

## Draw a Blue triangule around the dog face

```python
img = cv2.imread("DATA/dog_backpack.png")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

**Define vertices of the triangle and reshape them in the correct order for cv2**

```python
vertices = np.array([[250,700],[425,400],[600,700]],np.int32)
vertices.shape
```

```python
# Correct the order for cv2
pts = vertices.reshape((-1,1,2))
# Shape of the vertices must be (3, 1, 2)
pts.shape
```

**Draw the triangle***

```python
# isClosed to draw a closed polygon
cv2.polylines(img_rgb,[pts],isClosed=True,color=(0,0,255),thickness=20)
plt.imshow(img_rgb)
```

**Draw a filled polygon**

```python
img = cv2.imread("DATA/dog_backpack.png")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
cv2.fillPoly(img_rgb,[pts],color=(0,255,0))
plt.imshow(img_rgb)
```

## Display a window with the image and allow draw circle on the image with mouse right button

```python
import cv2
import numpy as np
# Create a function based on a CV2 Event (Left button click)
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),10)

# Open Image
img = cv2.imread("DATA/dog_backpack.png")
# This names the window so we can reference it 
cv2.namedWindow(winname='dog')
# Connects the mouse button to our callback function
cv2.setMouseCallback('dog',draw_circle)

while True: #Runs forever until we break with Esc key on keyboard
    # Shows the image window
    cv2.imshow('dog',img)
    # EXPLANATION FOR THIS LINE OF CODE:
    # https://stackoverflow.com/questions/35372700/whats-0xff-for-in-cv2-waitkey1/39201163
    if cv2.waitKey(20) & 0xFF == 27:
        break
# Once script is done, its usually good practice to call this line
# It closes all windows (just in case you have multiple windows called)
cv2.destroyAllWindows()
```
